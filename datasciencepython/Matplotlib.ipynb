```python
# demo matplotlib
import matplotlib.pyplot as plt

x = [10,20,30,40]
y = [20,30,40,50]

# plot data
plt.plot(x,y)

# assign title
plt.title("Simple Plot")

# label for x and y
plt.ylabel("y-axis")
plt.xlabel("x-axis")
plt.show()
```

```python
import pandas as pd
import seaborn as sns

data = pd.read_csv('employee.csv')
data.head()
```

```python
# univariate analysis for numerical value (Age)
sns.distplot(data['Age'])
```

```python
# univariate for categorical data (BusinessTravel)
sns.countplot(data['BusinessTravel'])
```

```python
# bivariate analysis categorical vs continuous 
sns.barplot(x=data['Department'],y=data['MonthlyIncome'])
```

```python
# bivariate continuous vs continuous (Age vs M.Income)
sns.scatterplot(x=data['Age'],y=data['MonthlyIncome'])
```
