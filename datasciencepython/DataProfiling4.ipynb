```python
# visualization
# common function:
# analyze() - analyze a single dataframe
# compare() - analyze 2 data frames
# compare_intra() - analyze 2 subsets of same dataframe
```

```python
# import libraries
import sweetviz as sv
import pandas as pd
```

```python
# read dataset
peng = pd.read_csv("penguins_cleaned.csv")
peng.tail()
```

```python
# discovering facts re species penguins
# y = species
# x = other info
# split dataframe
x = peng.drop('species',axis=1)
y = peng['species']
```

```python
# display x
x
```

```python
# display y
y
```

```python
# display all
peng
```

```python
# create an EDA report
# analyze whole penguins dataframe
analyze_report = sv.analyze(peng)
# create an html report file
# analyze_report.show_html('sv_analyze.html')

# create an html report, not open in browser
analyze_report.show_html('sv_analyze.html',open_browser=False)


```

```python
# x and y
# create a comparison report 
# Train vs Test | scikit-learn
# split x and y to test and train
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
# display x train dataframe
x_train
```

```python
# compare x_test vs x_train
compare_report = sv.compare([x_train,'Train'],[x_test,'Test'])
compare_report.show_html('sv_compare.html')
# compare_report.show_html('sv_compare.html',open_browser=False)
```
