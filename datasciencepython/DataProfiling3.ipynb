```python
# import libraries
import pandas as pd
import dabl
```

```python
# dataset: diabetes.csv
data = pd.read_csv('diabetes.csv')
data.head()
```

```python
# cleaning using dabl
cleandata = dabl.clean(data,target_col='Outcome',verbose=1)
cleandata.head()
```

```python
# temporarily convert a column to a diff feature
cleandata2 = dabl.clean(data, type_hints={"Pregnancies":"categorical"}, verbose=1)
cleandata2.head()
```

```python
# temporarily convert a column to a diff feature
cleandata2 = dabl.clean(data, type_hints=None, verbose=1)
cleandata2.head()
```

```python
# univariate and bivariate based on "Outcome"
dabl.plot(cleandata2, target_col="Outcome")
```
