```python
# reading a dataset from an online csv
ds_url = "https://raw.githubusercontent.com/fivethirtyeight/data/master/college-majors/recent-grads.csv"
data = pd.read_csv(ds_url)
data.sample()
```

```python
type(data)
```

```python
# x='Rank', y='Median', 'P25th', 'P75th'
data.plot(x='Rank',y=['P25th','Median','P75th'])
```

```python
# kinds
# area, bar, barh, box, hist, line, pie, scatter, hexbin, density
```

```python
# distributions
# each column can be used a series
median_col=data['Median']
# type(median_col)
median_col.plot(kind='hist')
```

```python
# sorting
top5 = data.sort_values(by='Median',ascending=False).head()
top5
```

```python
top5.plot(x='Major',y='Median',kind='bar',fontsize=6,rot=7)
```

```python
# asceding=True is set as default
top_medians = data[data['Median']>60000].sort_values("Median", ascending=False)
top_medians
```

```python
top_medians.plot(x="Major",y=['P25th','Median','P75th'],kind='bar',fontsize=5,rot=6)
```

```python
# correlation
# major with higher median = lower chance of unemployment?
data.plot(x="Median",y="Unemployment_rate", kind = 'scatter')
```

```python
# analyze categorical data using grouping
cat_totals=data.groupby("Major_category")["Total"].sum().sort_values(ascending=True)
cat_totals
# type(data.groupby("Major_category")["Total"])
# dir(data.groupby("Major_category")["Total"])
```

```python
# plot
cat_totals.plot(kind='barh',fontsize=4)
```

```python
# split of categories
small_cat_totals = cat_totals[cat_totals < 220_000]
big_cat_totals = cat_totals[cat_totals >= 220_000]

# merging series to form a new one
small_sums = pd.Series([small_cat_totals.sum()],index=["Other"])

# append to the series to be used
big_cat_totals = big_cat_totals.append(small_sums)

# append using pandas.concat
# big_cat_totals = pd.concat(big_cat_totals,small_sums)

# plot
big_cat_totals.plot(kind='pie',label='')
```
